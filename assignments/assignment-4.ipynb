```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp

import nengo
```

```python
from nengo.utils.ensemble import tuning_curves
from nengo.dists import Uniform

N = 100
tau_rc, tau_ref = 0.02, 0.002
lif_model = nengo.LIFRate(tau_rc=tau_rc, tau_ref=tau_ref)
model = nengo.Network(label='1D Neuron Ensemble')

with model:
    # Generate neurons
    neurons = nengo.Ensemble(N, dimensions=1,
                             max_rates=Uniform(100,200),
                             neuron_type = lif_model)
    # Generate decoders
    connection = nengo.Connection(neurons, neurons)#,
                                  #solver=nengo.solvers.LstsqNoise(noise=0.2))

# Simulate model
sim = nengo.Simulator(model)

# Plot tuning curves
x, A = tuning_curves(neurons, sim)

plt.plot(x, A)
plt.title('Neuron Ensemble')
plt.xlabel('$x$')
plt.ylabel('firing rate ($Hz$)');

# Plot representation
d = sim.data[connection].decoders.T
x, A_noisy = tuning_curves(neurons, sim)
#A_noisy = A + numpy.random.normal(scale=noise*numpy.max(A), size=A.shape)
xhat = np.dot(A_noisy, d)

print('RMSE with %d neurons is %g'%(N, np.sqrt(np.average((x-xhat)**2))))

plt.figure()
plt.plot(x, x)
plt.plot(x, xhat)
plt.xlabel('$x$')
plt.ylabel('$\hat{x}$')
plt.ylim(-1, 1)
plt.xlim(-1, 1);
```
